```python
import pandas as pd
```

```python
old = pd.read_csv("ncaa-womens-basketball-tournament-history.csv")
old
```

```python
finish = {
    "Champ": 1,
    "N2nd": 2,
    "NSF": 4,
    "RF": 8,
    "RSF": 16,
    "2nd": 32,
    "1st": 64,
    "OR": 32, # Just used in 1982
}
old["finish"] = old["Tourney finish"].apply(lambda x: finish[x])
old = old.rename(columns={"Year": "year", "School": "team", "Seed": "seed"})
old = old[["year", "team", "seed", "finish"]]
# Before 1984 the seeds are weird
old = old[old["year"] > 1983]
old
```

```python
new = pd.read_csv("recent_matches.csv")
new
```

```python
# Read in team name mappings
mapping = pd.read_csv("team_mapping.csv")
new["team"] = new["team"].apply(
    lambda x: mapping.loc[mapping["new"] == x].iloc[0]["old"]
    if x in list(mapping["new"])
    else x
)
new["opponent"] = new["opponent"].apply(
    lambda x: mapping.loc[mapping["new"] == x].iloc[0]["old"]
    if x in list(mapping["new"])
    else x
)
```

```python
# List of non-mapped teams. This should only include teams who debuted after 2016.
set(new["team"].unique()) - set(old["team"].unique())
```

```python
df = pd.concat([old, new])
df = df.sort_values(["year", "finish"])
df
```

```python
df.to_csv("all_matches.csv", index=False)
```
