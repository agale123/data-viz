```python
# Some rows didn't have enough commas when parsing the PDF to a csv so this pads
# the missing commas
file = open("2023_raw_raw.csv", "r")
lines = file.readlines()
file.close()


def fix_line(line):
    commas = line.count(",")
    if commas == 18:
        return line
    elif commas == 14:
        new_line = line.split(",")
        new_line.insert(8, ",,,")
        return ",".join(new_line)
    elif commas == 10:
        new_line = line.split(",")
        new_line.insert(4, ",,,,,,,")
        return ",".join(new_line)
    else:
        raise Exception("wrong comma count ", line.count(","), " for line ", line)


file = open("2023_raw.csv", "a")
file.writelines(list(map(fix_line, lines)))
file.close()
```
