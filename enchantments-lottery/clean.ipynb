```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("2023_raw.csv")
df
```

```python
df = df.drop(
    columns=[
        "Maximum Requested Group Size 1",
        "Maximum Requested Group Size 2",
        "Maximum Requested Group Size 3",
        "Processing",
        "State",
    ]
)
df = df.rename(
    columns={
        "Preferred Entry Date 1": "entry_date_1",
        "Preferred Division 1": "zone_1",
        "Minimum Acceptable Group Size 1": "group_size_1",
        "Preferred Entry Date 2": "entry_date_2",
        "Preferred Division 2": "zone_2",
        "Minimum Acceptable Group Size 2": "group_size_2",
        "Preferred Entry Date 3": "entry_date_3",
        "Preferred Division 3": "zone_3",
        "Minimum Acceptable Group Size 3": "group_size_3",
        "Results Status": "result",
        "Awarded Preference": "awarded_preference",
        "Awarded Entry Date": "awarded_entry_date",
        "Awarded Entrance Code/Name": "awarded_zone",
        "Awarded Group Size": "awarded_group_size",
    }
)
df
```

```python
def process_pref(df, i):
    # Just grab the columns for a specific preference
    df1 = df[
        [
            "entry_date_" + i,
            "zone_" + i,
            "group_size_" + i,
            "result",
            "awarded_preference",
        ]
    ]

    # Rename columns to match
    df1 = df1.rename(
        columns={
            "entry_date_" + i: "entry_date",
            "zone_" + i: "zone",
            "group_size_" + i: "group_size",
        }
    )

    # Only mark the status as successful if it was the selected preference
    df1.loc[df1["awarded_preference"] != int(i), "result"] = "Unsuccessful"
    
    # Mark as nan to remove requests after a successful one
    # df1.loc[df1["awarded_preference"] < int(i), "result"] = np.nan

    # Grab just the desired columns
    df1 = df1[["entry_date", "zone", "group_size", "result"]]

    # Remove na rows
    df1 = df1.dropna()
    
    df1 = df1.astype({"group_size": int})

    return df1

# Get the combined data
combined = pd.DataFrame([], columns = ["entry_date", "zone", "group_size", "result"])
for i in range(1, 4):
    temp = process_pref(df, str(i))
    combined = pd.concat([combined, temp])
    
combined = combined.reindex()

combined
```

```python
# Write to csv
combined.to_csv("2023.csv", index=False)
```
